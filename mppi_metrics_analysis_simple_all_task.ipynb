# MPPI Metrics Analysis — Learned Cost Shaping vs Default Weights

Implements **Metrics 1–4 and 6** from the Empirical Analysis Plan (§3–§6, §8 of the PDF).

**Data:** `eval_sweep_50_tray_push/npz/`  
Active batch sizes: **{BATCH_SIZES}** — 5 seed runs × 10 episodes each → **50 episodes per method** per batch size.  
_(To include all batch sizes [250, 500, 750, 1000] → 200 episodes per method, update `BATCH_SIZES` in the config cell.)_

| Metric | Coverage | Notes |
|--------|----------|-------|
| M1 — Entropy $\bar{H}_t$ | ✅ §3 | formula verified |
| M2 — nESS | ✅ §4 | formula verified |
| M3 — CV | ✅ §5 | formula verified |
| M4 — Max weight ratio $r_t$ | ✅ §6 | formula verified |
| M5 — Perf vs batch size | ↗ separate notebook | `benchmark_ARS_batch.ipynb` |
| M6 — Weight trajectories | ✅ §8 | |
| M7 — Variance decomposition | ❌ §9 | needs per-component rollout costs $C_i^{(j)}$ not saved in NPZ |

**Plot style note:** PDF requests "averaged + confidence bands" for M1–3; this notebook  
draws one curve per episode (more informative — episode spread visible directly).

**Notation (§2)**
- $J^{(j)}$ : total cost of rollout $j$ at one MPC step — `cem_cost_batch[ep][step, j]`
- $N$ : effective batch size per episode (N_batch or maxiter × N_batch depending on CEM_ITER_MODE)
- $\lambda = 10$ : MPPI temperature (`self.lamda` in `mjx_planner.py`)
- $\tilde{w}_j = \exp(-J^{(j)}/\lambda)$ unnormalised importance weight
- $\hat{w}_j = \tilde{w}_j / \sum_{j'} \tilde{w}_{j'}$ normalised importance weight

All four normalised metrics ($\bar{H}$, nESS, CV, $r_t$) divide by $N$ or $\log N$, so they are **comparable across batch sizes**.

**Pick / lift phase split (`task_flag`)**  
`task_flag[t] = 0` = pick phase; `task_flag[t] = 1` = move/lift phase.  
For pre-existing NPZ files: derived per episode from `cost_g_pick < 0.03` (`grab_pos_thresh`) and `cost_r_pick < 0.05` (`grab_rot_thresh`).  
The switch step is **different for every episode** — phase masks are computed per-episode.

## 1 — Load & merge sweep data

All 20 seed files per method (4 batch sizes × 5 seeds) are merged into one 200-episode dataset.

`cem_cost_batch` is kept in raw form: shape `(n_steps, maxiter_cem, N)` per episode.  
The next cell (`cd_cem_mode`) converts it to `cem_cost_batch` according to `CEM_ITER_MODE`:

| Mode | Shape of `cem_cost_batch[ep]` | N used in metrics |
|------|-------------------------------|-------------------|
| `"all"` | `(n_steps, maxiter_cem × N)` | N_eff = maxiter × N_batch |
| `"last"` | `(n_steps, N)` | N_batch (original batch size) |

To switch modes: change `CEM_ITER_MODE` in the config cell and re-run **only** `cd_cem_mode`.


```python
# ── Task-flag note ────────────────────────────────────────────────────────────
# `task_flag[t] = 0` during the approach phase (EEF approaching the object).
# `task_flag[t] = 1` during the post-switch phase (EEF engaging/manipulating the
#  object; labelled PHASE_AFTER — "lift" or "move" depending on TASK).
# Saved directly from state["task"] in every NPZ for the selected TASK.
# Thresholds that trigger the transition (for reference only):
#   grab_pos_thresh = 0.03  (both EEF positions must be < this)
#   grab_rot_thresh = 0.05  (both EEF rotations must be < this)
# ─────────────────────────────────────────────────────────────────────────────

GRAB_POS_THRESH = 0.03   # matches rl_trainer_batch.py: self.grab_pos_thresh
GRAB_ROT_THRESH = 0.05   # matches rl_trainer_batch.py: self.grab_rot_thresh


def load_sweep_dirs(dir_list):
    """
    Merge all seed NPZ files from a list of sweep sub-directories.
    cem_cost_batch is stored as (n_steps, maxiter_cem, N) per episode.
    The CEM-mode cell (cd_cem_mode) converts it to 2-D.
    """
    files = []
    for d in dir_list:
        files.extend(sorted(Path(d).glob("*.npz")))
    if not files:
        return None

    parts = [np.load(f, allow_pickle=True) for f in files]

    ep_1d  = ['success', 'total_time', 'reason']
    # Field names vary a little between tasks/dataset versions — list every
    # alias we've seen; they get normalised to one canonical name below.
    ep_obj = ['step_time', 'theta', 'thetadot',
              'cost_g_pick', 'cost_r_pick', 'cost_g_tray', 'cost_r_tray',
              'near_grasp', 'near_target', 'collision',
              'cost_weights', 'tray_target', 'task_flag',
              'cem_cost_batch',    # 3-D (n_steps, maxiter, N) per episode
              'cem_batch_costs']   # box_lift's alias for the same field
    meta   = ['batch_size', 'horizon', 'cost_weight_keys', 'cost_weights_keys']

    merged = {}
    for k in ep_1d:
        arrs = [p[k] for p in parts if k in p]
        if arrs:
            merged[k] = np.concatenate(arrs)
    for k in ep_obj:
        items = []
        for p in parts:
            if k in p:
                items.extend(p[k].tolist())
        if items:
            merged[k] = np.array(items, dtype=object)
    for k in meta:
        if k in parts[0]:
            merged[k] = parts[0][k]

    merged['batch_size_per_ep'] = np.array(
        [int(p['batch_size'][0]) for p in parts for _ in range(len(p['success']))]
    )

    # ── normalise field-name aliases to one canonical key ──────────────────────
    if 'cem_cost_batch' not in merged and 'cem_batch_costs' in merged:
        merged['cem_cost_batch'] = merged.pop('cem_batch_costs')
    else:
        merged.pop('cem_batch_costs', None)
    if 'cost_weight_keys' not in merged and 'cost_weights_keys' in merged:
        merged['cost_weight_keys'] = merged.pop('cost_weights_keys')
    else:
        merged.pop('cost_weights_keys', None)

    return merged, files


def ep_arr(obj_arr, ep, dtype=np.float64):
    return np.array(obj_arr[ep], dtype=dtype)


# ── load ───────────────────────────────────────────────────────────────────────
data = {}
for label, dir_list in SWEEP_DIRS.items():
    result = load_sweep_dirs(dir_list)
    if result is None:
        print(f"[SKIP] {label}: no files found")
        continue
    merged, files = result
    data[label] = merged
    sr = merged["success"].mean() * 100
    n  = len(merged["success"])
    raw_shape = np.array(merged['cem_cost_batch'][0]).shape   # (n_steps, maxiter, N)
    tf_source = "saved in NPZ"
    print(f"{label:10s}  n_ep={n}  SR={sr:.1f}%  ({len(files)} files)")
    print(f"  cem_cost_batch[ep0] shape : {raw_shape}  "
          f"(n_steps x maxiter_cem x N_batch)")
    print(f"  task_flag source          : {tf_source}")

labels      = list(data.keys())
WEIGHT_KEYS = [str(k) for k in data[labels[0]].get("cost_weight_keys", [])]
CEM_COMP_KEYS = []
print("\nRun the CEM-mode cell (cd_cem_mode) next to build cem_cost_batch.")
```

```python
# ── CEM iteration mode — re-run this cell whenever CEM_ITER_MODE changes ──────
#
# "all"  → pool all maxiter_cem rounds per step: shape (n_steps, maxiter × N)
#           N_eff = maxiter_cem × N_batch  (e.g. 3 × 250 = 750)
#
# "last" → final CEM iteration only: shape (n_steps, N)
#           Reflects the fully refined, most concentrated distribution.

def _build_cem_cost_batch(raw, mode):
    """raw: (n_steps, maxiter_cem, N) → 2-D array based on mode."""
    a = np.array(raw, dtype=np.float64)          # (n_steps, maxiter, N)
    if mode == "last":
        return a[:, -1, :]                        # (n_steps, N)
    else:                                         # "all"
        n_steps, maxiter, N = a.shape
        return a.reshape(n_steps, maxiter * N)    # (n_steps, maxiter*N)

for label in labels:
    d = data[label]
    if 'cem_cost_batch' not in d:
        print(f"[WARN] {label}: cem_cost_batch not found — skipping")
        continue
    # Preserve original 3-D (n_steps, maxiter, N) data before reshaping
    d['cem_batch_costs'] = d['cem_cost_batch']
    d['cem_cost_batch'] = np.array(
        [_build_cem_cost_batch(d['cem_batch_costs'][ep], CEM_ITER_MODE)
         for ep in range(len(d['cem_batch_costs']))],
        dtype=object,
    )

print(f"CEM_ITER_MODE = '{CEM_ITER_MODE}'")
print(f"{'Method':<12}  {'raw shape':>22}  →  {'cem_cost_batch[ep0]':>22}")
print("-" * 65)
for label in labels:
    d   = data[label]
    raw = np.array(d['cem_batch_costs'][0]).shape
    out = np.array(d['cem_cost_batch'][0]).shape
    n_steps, maxiter, N_base = raw
    N_eff = out[1]
    print(f"{label:<12}  {str(raw):>22}  →  {str(out):>22}"
          f"  (N_eff={N_eff})")
```

```python
# Sanity check: keys, shapes, task_flag switch distribution
print(f"WEIGHT_KEYS ({len(WEIGHT_KEYS)}): {WEIGHT_KEYS}")
print()
for label in labels:
    d   = data[label]
    n   = len(d['success'])
    cb0 = np.array(d['cem_cost_batch']).shape
    bs  = sorted(set(d['batch_size_per_ep'].tolist()))
    switches = []
    for ep in range(n):
        tf = ep_arr(d['task_flag'], ep)
        sw = np.where(tf >= 0.5)[0]
        switches.append(int(sw[0]) if len(sw) else -1)
    sw_valid = [s for s in switches if s >= 0]
    no_grab  = sum(1 for s in switches if s < 0)
    print(f"{label:10s}  n_ep={n}  SR={d['success'].mean()*100:.1f}%  batch_sizes={bs}")
    print(f"  cem_cost_batch[ep0] shape : {cb0}  (n_episodes)")
    if sw_valid:
        print(f"  task_flag switch step     : min={min(sw_valid)}  "
              f"median={int(np.median(sw_valid))}  max={max(sw_valid)}  "
              f"no-grab eps={no_grab}")
    print()
```

## Rollout Cost Distribution at Selected Time Instants

Steps are sampled from pick / transition / lift phases using the **mean switch step averaged across all episodes and all methods** being plotted.

- For each episode that successfully grabbed the tray, record `sw` = first `t` where `task_flag[t] ≥ 0.5`
- `switch = round(mean(sw))` across those episodes
- `pick_pool` = steps before `switch − TRANS_WINDOW`
- `trans_pool` = steps within `±TRANS_WINDOW` of `switch`
- `lift_pool` = steps after `switch + TRANS_WINDOW`

`N_PICK / N_TRANS / N_MOVE` steps are evenly sampled from each pool.

**Normalization — per episode, per method:**

$$\bar{c}^{(j)} = \frac{c^{(j)} - c^{\text{ep}}_{\min}}{c^{\text{ep}}_{\max} - c^{\text{ep}}_{\min}}$$

where $c^{\text{ep}}_{\min}$ and $c^{\text{ep}}_{\max}$ are the min and max costs across **all steps × all rollouts of that episode**.  
Each episode is normalised independently, and methods are normalised independently.  
Values near 0 = lowest-cost rollouts within that episode's range; near 1 = highest-cost.

Background colour: 🔵 pick  🟠 transition  🟢 lift


```python
import json as _json


def sample_pool(pool, n):
    if len(pool) == 0:
        return np.array([], dtype=int)
    idx = np.round(np.linspace(0, len(pool) - 1, min(n, len(pool)))).astype(int)
    return pool[idx]


# ════════════════════════════════════════════════════════════════════════════
# 1. STEP SELECTION — representative step indices for each phase (approach /
#    transition / move), based on the switch step averaged across every
#    episode and every method. Pure index selection — no cost data involved.
# ════════════════════════════════════════════════════════════════════════════

N_PICK       = 4    # steps sampled from approach phase
N_TRANS      = 2    # steps sampled from transition band
N_MOVE       = 4    # steps sampled from move phase
TRANS_WINDOW = 2    # half-width around switch defining "transition"

all_switches  = []
max_ep_len    = 0
per_method_sw = {}
for label in labels:
    d = data[label]
    sws = []
    for ep in range(len(d["success"])):
        task = ep_arr(d["task_flag"], ep)
        max_ep_len = max(max_ep_len, len(task))
        sw_idxs = np.where(task >= 0.5)[0]
        if len(sw_idxs):
            sws.append(int(sw_idxs[0]))
    per_method_sw[label] = sws
    all_switches.extend(sws)

assert all_switches, f"No episode grabbed the {TASK_OBJECT} — cannot determine average switch step"
switch  = int(round(np.mean(all_switches)))
ref_len = max_ep_len
n_grab  = len(all_switches)
n_total = sum(len(data[l]["success"]) for l in labels)

print("Switch step stats across all episodes:")
for label in labels:
    sws = per_method_sw[label]
    if sws:
        print(f"  {label:<12}  grabbed={len(sws)}  "
              f"min={min(sws)}  mean={int(round(np.mean(sws)))}  max={max(sws)}")
    else:
        print(f"  {label:<12}  grabbed=0 (no eps)")
print(f"  {'Overall':<12}  grabbed={n_grab}/{n_total}  "
      f"mean switch={switch}  range=[{min(all_switches)}, {max(all_switches)}]")
print(f"  max episode length: {ref_len}")

t_pick_end   = max(switch - TRANS_WINDOW - 1, 0)
t_move_start = switch + TRANS_WINDOW + 1

pick_pool  = np.arange(0, t_pick_end + 1)
trans_pool = np.arange(max(switch - TRANS_WINDOW, 0),
                       min(switch + TRANS_WINDOW + 1, ref_len))
move_pool  = np.arange(t_move_start, ref_len)

selected_pick  = sample_pool(pick_pool,  N_PICK)
selected_trans = sample_pool(trans_pool, N_TRANS)
selected_move  = sample_pool(move_pool,  N_MOVE)
selected = np.concatenate([selected_pick, selected_trans, selected_move])
phases   = ([PHASE_BEFORE]   * len(selected_pick)
          + ["transition"] * len(selected_trans)
          + [PHASE_AFTER]    * len(selected_move))
print(f"\nSelected steps: pick={selected_pick.tolist()}  "
      f"trans={selected_trans.tolist()}  move={selected_move.tolist()}")
```

```python
# ════════════════════════════════════════════════════════════════════════════
# 2. NORMALIZATION — the ONLY place raw rollout costs get normalised.
#    Every (method, episode, step, cem_iter) is min-max scaled independently,
#    using just the N rollouts at that (step, iter) — the batch axis is the
#    only one reduced over. CEM_ITER_MODE then pools iterations together
#    ("all") or keeps just the last one ("last") — a reshape, not a
#    renormalisation.
# ════════════════════════════════════════════════════════════════════════════

def normalize_per_step_iter(raw):
    """raw: (n_steps, maxiter, N) -> min-max normalised, same shape.
    min/max taken over N (last axis) for every (step, iter) independently."""
    s_min = raw.min(axis=-1, keepdims=True)
    s_max = raw.max(axis=-1, keepdims=True)
    denom = s_max - s_min
    return (raw - s_min) / denom

def pool_cem_iters(norm, mode):
    """norm: (n_steps, maxiter, N) normalised -> (n_steps, N_eff)."""
    if mode == "all":
        n_steps, maxiter, N = norm.shape
        return norm.reshape(n_steps, maxiter * N)
    return norm[:, -1, :]   # "last"

# ── overview: conceptual (method, episode, step, maxiter) shape ───────────────
#    step count and N can vary per episode, so this isn't one rectangular array —
#    printed as ranges per method instead.
print("Raw cost tensor per (method, episode): (n_steps, maxiter, N)")
print(f"  {'Method':<12}  {'n_ep':>5}  {'n_steps range':>16}  {'maxiter':>8}  {'N':>6}")
for l in labels:
    d = data[l]
    n_ep = len(d["success"])
    step_counts = [np.asarray(d["cem_batch_costs"][ep]).shape[0] for ep in range(n_ep)]
    sample = np.asarray(d["cem_batch_costs"][0])
    maxiter_l, N_l = sample.shape[1], sample.shape[2]
    print(f"  {l:<12}  {n_ep:>5}  [{min(step_counts):>5}, {max(step_counts):>5}]     "
          f"{maxiter_l:>8}  {N_l:>6}")
print(f"\nConceptual shape: (method={len(labels)}, episode~{n_ep} each, "
      f"step=ragged per episode, maxiter={maxiter_l})\n")

ep_norm_J = {}   # ep_norm_J[l][ep] : (n_steps, N_eff) fully normalised — the single source of truth
ep_stats  = {}   # ep_stats[l][ep]  : raw min/max/mean summary, for printing + JSON export only

_traced = False   # print a one-shot shape trace on the very first (method, ep) only

for l in labels:
    d = data[l]
    ep_norm_J[l] = []
    ep_stats[l]  = []
    for ep in range(len(d["success"])):
        raw = np.array(d["cem_batch_costs"][ep], dtype=np.float64)  # (n_steps, maxiter, N)
        maxiter = raw.shape[1]

        normed = normalize_per_step_iter(raw)
        pooled = pool_cem_iters(normed, CEM_ITER_MODE)
        ep_norm_J[l].append(pooled)

        if not _traced:
            _traced = True
            n_steps, _, N = raw.shape
            print(f"Shape pipeline (worked example: method={l!r}, ep={ep}):\n")
            print(f"  conceptual full tensor         (method, ep, step, maxiter, batch)  "
                  f"~ ({len(labels)}, n_ep, ragged, {maxiter}, {N})")
            print(f"  raw[method][ep]                (step, maxiter, batch)               = {raw.shape}")
            print(f"  normalize_per_step_iter(raw)   (step, maxiter, batch)               = {normed.shape}"
                  "   [min/max over batch axis only, per (step, maxiter)]")
            print(f"  pool_cem_iters(..., {CEM_ITER_MODE!r})       (step, batch_eff)                    = {pooled.shape}"
                  "   [maxiter folded into batch axis]")
            print(f"  ep_norm_J[method][ep][0]       (batch_eff,)                         = {pooled[0].shape}"
                  "   <- one planning step, all normalised rollouts\n")

        iters = [{"iter": it,
                  "min":  round(float(raw[:, it, :].min()),  4),
                  "max":  round(float(raw[:, it, :].max()),  4),
                  "mean": round(float(raw[:, it, :].mean()), 4)}
                 for it in range(maxiter)]
        ep_stats[l].append({
            "ep":      ep,
            "min":     round(float(raw.min()),  4),
            "max":     round(float(raw.max()),  4),
            "mean":    round(float(raw.mean()), 4),
            "success": bool(d["success"][ep]),
            "iters":   iters,
        })

print(f"\n ep_norm_J, {np.size(ep_norm_J)}")

# print("\nPer-episode cost stats (raw) with per-CEM-iteration breakdown:")
# print(f"  {'Method':<12}  {'ep':>4}  {'ok':>3}  "
#       f"{'ep_min':>10}  {'ep_max':>14}  {'ep_mean':>12}  | iters (min / max)")
# print("  " + "-" * 80)
for l in labels:
    for s in ep_stats[l]:
        iter_str = "  ".join(
            f"it{r['iter']}:[{r['min']:.1f},{r['max']:.1f}]"
            for r in s["iters"]
        )
        print(f"  {l:<12}  {s['ep']:>4}  {'✓' if s['success'] else '✗':>3}  "
              f"{s['min']:>10.2f}  {s['max']:>14.2f}  {s['mean']:>12.2f}  | {iter_str}")
    print()

# json_path = "ep_cost_stats.json"
# with open(json_path, "w") as f:
#     _json.dump({l: ep_stats[l] for l in labels}, f, indent=2)
# print(f"Saved → {json_path}")


# ════════════════════════════════════════════════════════════════════════════
# 3. Collect normalised rollout costs at each selected step
# ════════════════════════════════════════════════════════════════════════════

phase_colors = {PHASE_BEFORE: "#cce5ff", "transition": "#ffe0b2", PHASE_AFTER: "#c8e6c9"}

norm_costs  = {l: [] for l in labels}
ep_med_data = {l: [] for l in labels}
n_ep_used   = {l: [] for l in labels}

for l in labels:
    d  = data[l]
    ok = d["success"].astype(bool)
    for t in selected:
        pool = []
        meds = []
        for ep in range(len(d["success"])):
            J_norm = ep_norm_J[l][ep]          # (n_steps, N_eff) — already normalised
            if t >= len(J_norm):
                continue
            normed = J_norm[t]                 # (N_eff,)
            pool.append(normed)
            meds.append((float(np.median(normed)), bool(ok[ep])))
        norm_costs[l].append(np.concatenate(pool) if pool else np.array([np.nan]))
        ep_med_data[l].append(meds)
        n_ep_used[l].append(len(meds))
```

```python
# ════════════════════════════════════════════════════════════════════════════
# Per-episode-phase pooling — comprises ALL episodes.
# No new normalisation: reuses ep_norm_J from the cell above as-is.
#
# Each episode has its OWN switch step (when task_flag first hits 1). Every
# episode's own phase span is computed from that switch step — not a
# population-level mean/CI margin — so no episode is ever dropped for being
# "unsafe":
#   PHASE_BEFORE : steps in [0, sw_ep - TW)
#   transition   : steps in [sw_ep - TW, sw_ep + TW + 1)
#   PHASE_AFTER  : steps in [sw_ep + TW + 1, n_steps)
# Within each phase span, linspace_steps evenly spaces N_PICK/N_TRANS/N_MOVE
# sample points across the WHOLE span, so a long post-switch phase and a short
# one contribute comparably instead of the longer one dominating by step count.
# Episodes that never grab (no switch step) are INCLUDED, not skipped: their
# entire length counts as PHASE_BEFORE (they never reached transition/after —
# those phases genuinely have no data from that episode, which is correct).
# ════════════════════════════════════════════════════════════════════════════

def linspace_steps(start, end, n):
    """n evenly-spaced integer step indices spanning [start, end)."""
    if end <= start:
        return np.array([], dtype=int)
    idx = np.round(np.linspace(start, end - 1, min(n, end - start))).astype(int)
    return np.unique(idx)

TW = TRANS_WINDOW  # reuse the same half-width used for step selection above

phase_pool = {l: {PHASE_BEFORE: [], "transition": [], PHASE_AFTER: []} for l in labels}

for l in labels:
    d = data[l]
    for ep in range(len(d["success"])):
        task    = ep_arr(d["task_flag"], ep)
        sw_idxs = np.where(task >= 0.5)[0]
        J_norm  = ep_norm_J[l][ep]        # (n_steps, N_eff) — already normalised
        n_steps = len(J_norm)

        if len(sw_idxs) == 0:
            # never grabbed -> whole episode is PHASE_BEFORE only
            approach_steps = linspace_steps(0, n_steps, N_PICK)
            if len(approach_steps):
                phase_pool[l][PHASE_BEFORE].append(J_norm[approach_steps].reshape(-1))
            continue
        sw_ep = int(sw_idxs[0])

        approach_end = max(sw_ep - TW, 0)
        trans_start  = approach_end
        trans_end    = min(sw_ep + TW + 1, n_steps)
        move_start   = trans_end

        approach_steps = linspace_steps(0, approach_end, N_PICK)
        trans_steps    = linspace_steps(trans_start, trans_end, N_TRANS)
        move_steps     = linspace_steps(move_start, n_steps, N_MOVE)

        if len(approach_steps):
            phase_pool[l][PHASE_BEFORE].append(J_norm[approach_steps].reshape(-1))
        if len(trans_steps):
            phase_pool[l]["transition"].append(J_norm[trans_steps].reshape(-1))
        if len(move_steps):
            phase_pool[l][PHASE_AFTER].append(J_norm[move_steps].reshape(-1))

PHASES = (PHASE_BEFORE, "transition", PHASE_AFTER)

for l in labels:
    for ph in PHASES:
        phase_pool[l][ph] = (np.concatenate(phase_pool[l][ph])
                              if phase_pool[l][ph] else np.array([np.nan]))

print(f"{'Method':<12}  {PHASE_BEFORE + ' n':>11}  {'transition n':>13}  {PHASE_AFTER + ' n':>8}")
for l in labels:
    print(f"  {l:<12}  {len(phase_pool[l][PHASE_BEFORE]):>11}  "
          f"{len(phase_pool[l]['transition']):>13}  {len(phase_pool[l][PHASE_AFTER]):>8}")


# ── plot: one violin per method, per phase — 3 phases side by side ────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 5.5))
for ax, ph in zip(axes, PHASES):
    plot_data = [phase_pool[l][ph] for l in labels]
    parts = ax.violinplot(plot_data, positions=range(len(labels)),
                          showmedians=True, showextrema=True)
    for pc, l in zip(parts["bodies"], labels):
        pc.set_facecolor(COLORS.get(l, "gray"))
        pc.set_alpha(0.7)
    for part in ("cmedians", "cmins", "cmaxes", "cbars"):
        parts[part].set_color("black")
        parts[part].set_linewidth(1.2)

    ax.axhline(LOW_THRESH, color="red", lw=1.2, ls="--", alpha=0.7)

    for xi, l in enumerate(labels):
        vals = phase_pool[l][ph]
        vals = vals[~np.isnan(vals)]
        pct  = 100.0 * np.mean(vals < LOW_THRESH) if len(vals) else float("nan")
        ax.text(xi, -0.06, f"{pct:.0f}%\n(n={len(vals)})",
                ha="center", va="top", fontsize=10, fontweight="bold",
                color=COLORS.get(l, "black"), transform=ax.get_xaxis_transform())

    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, fontsize=12)
    ax.set_ylim(-0.12, 1.05)
    ax.set_title(ph, fontsize=14, fontweight="bold")
    ax.grid(axis="y", lw=0.4, color="gray", alpha=0.4)

axes[0].set_ylabel("Normalised rollout cost", fontsize=12)
fig.suptitle(
    f"Rollout cost distribution per phase — {TASK} — all steps × all episodes, "
    "phase assigned per episode's own switch step\n"
    f"(% below threshold {LOW_THRESH} · n = total rollout-step samples pooled)",
    fontsize=13, fontweight="bold")
plt.tight_layout()
phase_pool_img = f"phase_pool_{TASK}.png"
fig.savefig(phase_pool_img, dpi=150, bbox_inches="tight")
print(f"Saved → {phase_pool_img}")
plt.show()
```

```python
# ════════════════════════════════════════════════════════════════════════════
# nESS (normalized Effective Sample Size) per phase — explains *sample
# efficiency*, not just cost level. Matches the actual production weighting
# (compute_adaptive_mean_cov): weights = softmax(-costs * self.lamda) — i.e.
# LAMBDA MULTIPLIES the cost (inverse-temperature convention), not divides
# it as in the paper's exp(-c/λ) notation:
#   ŵ_j = exp(-(c_j - min(c)) * LAMBDA) / Σ_j' exp(-(c_j' - min(c)) * LAMBDA)
#   nESS = 1 / (N_eff · Σ ŵ_j²)                      (∈ (0, 1], 1 = all N
#                                                       rollouts equally useful)
# Computed from RAW rollout costs (d["cem_cost_batch"], already reduced per
# CEM_ITER_MODE) — not the per-group min-max normalised ep_norm_J — since
# nESS only needs relative cost differences, which min-max normalisation
# doesn't change but which the constant shift by min(c) makes numerically
# stable without altering the result.
#
# Restricted to BATCH_SIZES=[250] — the sample-constrained regime where the
# sample-efficiency claim (fewer wasted rollouts ⇒ better performance at
# small N) is most directly testable. Uses the SAME per-episode phase
# boundaries as cd_phase_pool, but — unlike cd_phase_pool's linspace_steps
# subsampling — includes EVERY step within each phase: nESS is already a
# cheap per-step scalar (no rollout-level cost to pool), so there's no need
# to subsample for balance the way the per-rollout violins do. Episodes
# that never grab are INCLUDED (their whole length counts as PHASE_BEFORE),
# not skipped.
# ════════════════════════════════════════════════════════════════════════════

assert BATCH_SIZES == [250], (
    f"nESS analysis is set up for BATCH_SIZES=[250] (the sample-constrained "
    f"regime); got {BATCH_SIZES}. Set BATCH_SIZES = [250] in the config cell "
    f"and re-run from cd_83825 if you want this analysis."
)

TW = TRANS_WINDOW  # defined locally — don't rely on cd_phase_pool having run

def step_ness(J):
    """J : (N_eff,) raw rollout costs at one step -> scalar nESS in (0, 1].
    weights = softmax(-J * LAMBDA), matching compute_adaptive_mean_cov."""
    w     = np.exp(-(J - J.min()) * LAMBDA)
    w_hat = w / w.sum()
    return 1.0 / (len(J) * np.sum(w_hat ** 2))

ness_pool = {l: {PHASE_BEFORE: [], "transition": [], PHASE_AFTER: []} for l in labels}

for l in labels:
    d = data[l]
    for ep in range(len(d["success"])):
        task    = ep_arr(d["task_flag"], ep)
        sw_idxs = np.where(task >= 0.5)[0]
        raw_ep  = np.array(d["cem_cost_batch"][ep], dtype=np.float64)  # (n_steps, N_eff) RAW
        n_steps = len(raw_ep)

        if len(sw_idxs) == 0:
            # never grabbed -> whole episode is PHASE_BEFORE only
            for t in range(n_steps):
                ness_pool[l][PHASE_BEFORE].append(step_ness(raw_ep[t]))
            continue
        sw_ep = int(sw_idxs[0])

        approach_end = max(sw_ep - TW, 0)
        trans_start  = approach_end
        trans_end    = min(sw_ep + TW + 1, n_steps)
        move_start   = trans_end

        for t in range(0, approach_end):
            ness_pool[l][PHASE_BEFORE].append(step_ness(raw_ep[t]))
        for t in range(trans_start, trans_end):
            ness_pool[l]["transition"].append(step_ness(raw_ep[t]))
        for t in range(move_start, n_steps):
            ness_pool[l][PHASE_AFTER].append(step_ness(raw_ep[t]))

for l in labels:
    for ph in PHASES:
        ness_pool[l][ph] = (np.array(ness_pool[l][ph])
                             if ness_pool[l][ph] else np.array([np.nan]))

print(f"nESS per phase (N={BATCH_SIZES[0]}, mean ± std across sampled steps × episodes):")
print(f"  {'Method':<12}  {PHASE_BEFORE:>14}  {'transition':>14}  {PHASE_AFTER:>14}")
for l in labels:
    row = "  ".join(f"{np.nanmean(ness_pool[l][ph]):>7.3f} ± {np.nanstd(ness_pool[l][ph]):.3f}"
                     for ph in PHASES)
    print(f"  {l:<12}  {row}")


# ── plot: one violin per method, per phase — 3 phases side by side ────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 5.5))
for ax, ph in zip(axes, PHASES):
    plot_data = [ness_pool[l][ph] for l in labels]
    parts = ax.violinplot(plot_data, positions=range(len(labels)),
                          showmedians=True, showextrema=True)
    for pc, l in zip(parts["bodies"], labels):
        pc.set_facecolor(COLORS.get(l, "gray"))
        pc.set_alpha(0.7)
    for part in ("cmedians", "cmins", "cmaxes", "cbars"):
        parts[part].set_color("black")
        parts[part].set_linewidth(1.2)

    for xi, l in enumerate(labels):
        vals = ness_pool[l][ph]
        vals = vals[~np.isnan(vals)]
        mean_val = np.mean(vals) if len(vals) else float("nan")
        ax.text(xi, -0.06, f"μ={mean_val:.2f}\n(n={len(vals)})",
                ha="center", va="top", fontsize=10, fontweight="bold",
                color=COLORS.get(l, "black"), transform=ax.get_xaxis_transform())

    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, fontsize=12)
    ax.set_ylim(-0.12, 1.05)
    ax.set_title(ph, fontsize=14, fontweight="bold")
    ax.grid(axis="y", lw=0.4, color="gray", alpha=0.4)

axes[0].set_ylabel(f"nESS  (N={BATCH_SIZES[0]})", fontsize=12)
fig.suptitle(
    f"Normalized Effective Sample Size per phase — {TASK} (N={BATCH_SIZES[0]}) — "
    "phase assigned per episode's own switch step\n"
    "Higher nESS = more of the rollout batch meaningfully contributes to the MPPI update",
    fontsize=13, fontweight="bold")
plt.tight_layout()
ness_img = f"ness_phase_{TASK}.png"
fig.savefig(ness_img, dpi=150, bbox_inches="tight")
print(f"Saved → {ness_img}")
plt.show()
```

```python
# ════════════════════════════════════════════════════════════════════════════
# nESS vs. episode outcome — WITHIN-method correlation.
#
# Comparing nESS magnitude ACROSS methods (previous cell) is ambiguous: low
# nESS could mean "sharply concentrated on genuinely good rollouts" (good) or
# "degenerate, arbitrary few survivors" (bad) — you can't tell which from the
# number alone. This cell asks a more direct question that sidesteps that
# ambiguity: for a SINGLE method, do its own successful episodes have
# different (mean per-episode) nESS than its own failed episodes? This
# controls for the method entirely — it only compares a method against
# itself, split by outcome — so if higher nESS predicts success within the
# same method, that's much harder to dismiss as some other confound of
# "which method" than a cross-method nESS comparison.
#
# For each (method, episode, phase): average step_ness over every step of
# that episode within that phase -> one scalar "episode nESS" per phase.
# Then split those scalars by d["success"][ep] and compare the two groups
# (mean difference + Mann-Whitney U p-value, since nESS is not normally
# distributed).
# ════════════════════════════════════════════════════════════════════════════

from scipy.stats import mannwhitneyu

ep_ness    = {l: {ph: [] for ph in PHASES} for l in labels}   # per-episode mean nESS
ep_outcome = {l: {ph: [] for ph in PHASES} for l in labels}   # aligned bool success flag

for l in labels:
    d = data[l]
    for ep in range(len(d["success"])):
        task    = ep_arr(d["task_flag"], ep)
        sw_idxs = np.where(task >= 0.5)[0]
        raw_ep  = np.array(d["cem_cost_batch"][ep], dtype=np.float64)
        n_steps = len(raw_ep)
        ok      = bool(d["success"][ep])

        if len(sw_idxs) == 0:
            vals = [step_ness(raw_ep[t]) for t in range(n_steps)]
            if vals:
                ep_ness[l][PHASE_BEFORE].append(np.mean(vals))
                ep_outcome[l][PHASE_BEFORE].append(ok)
            continue
        sw_ep = int(sw_idxs[0])

        approach_end = max(sw_ep - TW, 0)
        trans_start  = approach_end
        trans_end    = min(sw_ep + TW + 1, n_steps)
        move_start   = trans_end

        for ph, lo, hi in ((PHASE_BEFORE, 0, approach_end),
                           ("transition", trans_start, trans_end),
                           (PHASE_AFTER, move_start, n_steps)):
            vals = [step_ness(raw_ep[t]) for t in range(lo, hi)]
            if vals:
                ep_ness[l][ph].append(np.mean(vals))
                ep_outcome[l][ph].append(ok)

for l in labels:
    for ph in PHASES:
        ep_ness[l][ph]    = np.array(ep_ness[l][ph])
        ep_outcome[l][ph] = np.array(ep_outcome[l][ph], dtype=bool)

print(f"Per-episode mean nESS: success vs fail, WITHIN method (N={BATCH_SIZES[0]})")
print(f"  {'Method':<12} {'Phase':<12} {'n_ok':>5} {'n_fail':>6}  {'nESS|ok':>9}  {'nESS|fail':>9}  {'diff':>8}  {'p (MWU)':>9}")
print("  " + "-" * 80)
for l in labels:
    for ph in PHASES:
        ok_vals   = ep_ness[l][ph][ep_outcome[l][ph]]
        fail_vals = ep_ness[l][ph][~ep_outcome[l][ph]]
        m_ok   = np.mean(ok_vals)   if len(ok_vals)   else float("nan")
        m_fail = np.mean(fail_vals) if len(fail_vals) else float("nan")
        if len(ok_vals) >= 2 and len(fail_vals) >= 2:
            p = mannwhitneyu(ok_vals, fail_vals, alternative="two-sided").pvalue
            p_str = f"{p:.3f}"
        else:
            p_str = "n/a"
        print(f"  {l:<12} {ph:<12} {len(ok_vals):>5} {len(fail_vals):>6}  "
              f"{m_ok:>9.3f}  {m_fail:>9.3f}  {m_ok - m_fail:>+8.3f}  {p_str:>9}")
    print()


# ── plot: grouped bars (success vs fail), one panel per phase ─────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 5.5))
bar_w = 0.35
x = np.arange(len(labels))
for ax, ph in zip(axes, PHASES):
    ok_means, ok_sems, fail_means, fail_sems = [], [], [], []
    for l in labels:
        ok_vals   = ep_ness[l][ph][ep_outcome[l][ph]]
        fail_vals = ep_ness[l][ph][~ep_outcome[l][ph]]
        ok_means.append(np.mean(ok_vals) if len(ok_vals) else np.nan)
        ok_sems.append(np.std(ok_vals) / np.sqrt(len(ok_vals)) if len(ok_vals) > 1 else 0.0)
        fail_means.append(np.mean(fail_vals) if len(fail_vals) else np.nan)
        fail_sems.append(np.std(fail_vals) / np.sqrt(len(fail_vals)) if len(fail_vals) > 1 else 0.0)

    ax.bar(x - bar_w / 2, ok_means, bar_w, yerr=ok_sems, capsize=4,
           label="success", color="#2ca02c", alpha=0.85, edgecolor="k", linewidth=0.6)
    ax.bar(x + bar_w / 2, fail_means, bar_w, yerr=fail_sems, capsize=4,
           label="fail", color="#d62728", alpha=0.85, edgecolor="k", linewidth=0.6)

    for xi, l in enumerate(labels):
        n_ok   = int(np.sum(ep_outcome[l][ph]))
        n_fail = len(ep_outcome[l][ph]) - n_ok
        ax.text(xi, -0.11, f"n={n_ok}/{n_fail}", ha="center", va="top",
                fontsize=9, transform=ax.get_xaxis_transform())

    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=12)
    ax.set_title(ph, fontsize=14, fontweight="bold")
    ax.grid(axis="y", lw=0.4, color="gray", alpha=0.4)

axes[0].set_ylabel("mean per-episode nESS", fontsize=12)
axes[0].legend(fontsize=10, loc="upper right")
fig.suptitle(
    f"nESS vs. episode outcome, within method — {TASK} (N={BATCH_SIZES[0]})\n"
    "success/fail compared only against the SAME method's own episodes "
    "(n = n_success/n_fail per bar)",
    fontsize=13, fontweight="bold")
plt.tight_layout()
corr_img = f"ness_success_corr_{TASK}.png"
fig.savefig(corr_img, dpi=150, bbox_inches="tight")
print(f"Saved → {corr_img}")
plt.show()
```

```python
# ── Per-method rollout cost violins with per-method phase boundaries ───────────
# Problem with the shared-switch plot: step t=45 may be pick for ARS
# (whose avg switch=55) but move for Hand-tuned (avg switch=40).
# Fix: each method row selects its own representative steps using that
# method's own median switch step. Approach and move steps are linspace-
# sampled (via linspace_steps, defined in the phase-pooling cell above)
# across the WHOLE span of each phase — [0, m_sw-TW) and [m_sw+TW+1, m_ref) —
# no CI margin, so a method's approach/move phase can never collapse the way
# it did with a CI-restricted "safe" window.

_N_PICK  = 4
_N_TRANS = 1
_N_MOVE  = 2
_TW      = 1   # transition-window half-width (kept tight around the switch)
_GAP     = 6   # extra buffer so the LAST approach / FIRST move sample stays
                # clearly separated from the transition sample, not adjacent to it
_N_TOTAL = _N_PICK + _N_TRANS + _N_MOVE

# ── per-method step selection based on that method's own avg switch ────────────
method_switch    = {}   # per-method average switch step
method_selected  = {}   # per-method selected step indices
method_phases    = {}   # per-method phase labels
method_nc        = {}   # per-method normalized costs at selected steps
method_bounds    = {}   # per-method (approach_end, m_sw, move_start)

for l in labels:
    sws = per_method_sw.get(l, [])
    print(f"\nMethod {l} switch steps:  min={min(sws)}  median={int(round(np.median(sws))) if sws else 'N/A'}  max={max(sws)}  n={len(sws)}")
    m_sw  = int(round(np.median(sws))) if sws else switch   # fallback to global
    method_switch[l] = m_sw

    m_ref = max(len(ep_norm_J[l][ep]) for ep in range(len(data[l]["success"])))

    approach_end = max(m_sw - _GAP, 0)
    move_start   = m_sw + _GAP
    print(f"  approach_end={approach_end}  m_sw={m_sw}  move_start={move_start}")
    method_bounds[l] = (approach_end, m_sw, move_start)

    sel_p  = linspace_steps(0, approach_end, _N_PICK)
    t_pool = np.arange(max(m_sw - _TW, 0), min(m_sw + _TW + 3, m_ref))
    sel_t  = sample_pool(t_pool, _N_TRANS)
    sel_m  = linspace_steps(move_start, m_ref, _N_MOVE)
    sel    = np.concatenate([sel_p, sel_t, sel_m])
    phs    = ([PHASE_BEFORE]   * len(sel_p)
           + ["transition"] * len(sel_t)
           + [PHASE_AFTER]    * len(sel_m))

    method_selected[l] = sel
    method_phases[l]   = phs

    # collect normalised costs
    d  = data[l]

    nc = []
    for t in sel:
        pool = []
        for ep in range(len(d["success"])):
            J_norm = ep_norm_J[l][ep]
            if t >= len(J_norm):
                continue
            pool.append(J_norm[t])
        nc.append(np.concatenate(pool) if pool else np.array([np.nan]))
    method_nc[l] = nc


# ── Phase boundary table ───────────────────────────────────────────────────────
print(f"\n{'Method':<14} {'median_sw':>9} {'approach_end':>13} {'move_start':>11}")
print("-" * 55)
for l in labels:
    ae, msw, ms = method_bounds[l]
    print(f"  {l:<12}  {msw:>8}  {ae:>13}  {ms:>11}")
print("-" * 55)
print("  approach steps ∈ [0, approach_end]  |  move steps ∈ [move_start, ep_end]  "
      "(linspace across full span, no CI margin)")

print("\nSelected step indices per method:")
for l in labels:
    print(f"  {l:<12}  switch≈{method_switch[l]}  "
          f"pick={[x for x,p in zip(method_selected[l], method_phases[l]) if p==PHASE_BEFORE]}  "
          f"transition={[x for x,p in zip(method_selected[l], method_phases[l]) if p=='transition']}  "
          f"move={[x for x,p in zip(method_selected[l], method_phases[l]) if p==PHASE_AFTER]}")
```
